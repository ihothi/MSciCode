```python
import numpy as np
from astropy.io import fits
import matplotlib
import matplotlib.pyplot as plt
```

```python
class Quasar:
    
    def __init__():[]
    def __init__(self, SDSS_NAME, R, D, PSFFLUX, PSFMAG, Z_VI):
        #leaving out redshift for now
        self.name = SDSS_NAME
        self.RA = R
        self.Dec = D
        self.flux = PSFFLUX
        self.Mag = PSFMAG
        self.z = Z_VI
```

```python
list2 = fits.open('Superset_DR12Q.fits',memmap=True)#opening file
super=list2[1].data # storing  BINTABLE extension data
CurrentObject = 0
xstar=[]
xgal=[]
xqso=[]
ystar=[]
ygal=[]
yqso=[]
while CurrentObject < len(super):
    Current = super[CurrentObject]
    if Current['CLASS_PERSON'] == 1:
        xstar.append(Current['RA'])
        ystar.append(Current['Dec'])
    
    elif Current['CLASS_PERSON'] == 3:
        xqso.append(Current['RA'])
        yqso.append(Current['Dec'])
        
    elif Current['CLASS_PERSON'] == 4:
        xgal.append(Current['RA'])
        ygal.append(Current['Dec'])

    CurrentObject=CurrentObject+1
```

```python
plt.plot(xqso,yqso, 'ro')

plt.plot(xstar,ystar, 'yo')
plt.plot(xgal,ygal, 'go')

plt.title('Object positions')
plt.xlabel('RA')
plt.ylabel('Dec')
#plt.xlim(141,142)
plt.legend(['Quasars','Stars', 'Galaxies' ], loc='upper left')
plt.show()
```

# Plate

```python
list = fits.open('spPlate-4564-55570.fits',memmap=True)#opening file
```

```python
Bin_info = list[5].data
i=0
x=[]
y=[]

while i < len(Bin_info):
    plate_quasar = Bin_info[i]
    x.append(round(plate_quasar['RA'],2))
    y.append(round(plate_quasar['Dec'],2))
    i=i+1
```

```python
plt.title('Object positions')
plt.xlabel('RA')
plt.ylabel('Dec')
plt.plot(x,y, 'o')
plt.show()
```
