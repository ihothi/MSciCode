```python
class Quasar:
    
    def __init__():[]
    def __init__(self, SDSS_NAME, R, D, PSFFLUX, PSFMAG, Z_VI):
        #leaving out redshift for now
        self.name = SDSS_NAME
        self.RA = R
        self.Dec = D
        self.flux = PSFFLUX
        self.Mag = PSFMAG
        self.z = Z_VI
      
    def QuasarInfo(self):
        return [self.name, self.RA, self.Dec, self.flux,self.Mag,self.z]
   
        
```
